```python
import matplotlib.pyplot as plt
from collections import defaultdict
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline

sns.set()
df = pd.read_csv("csvs/all_data.csv")
```

```python
### HIERARCHY = '[2]' ###


sns_plot = sns.relplot(data=df[df.hierarchy=='[2]'], x="request_count", y="runtime", style="k_means_seed", kind="line")
sns_plot.set(xlabel='Amount of VNF requests', ylabel='Runtime of approach in seconds')
sns_plot.savefig("figs/hierarchy_2_runtimes.pdf")
sns_plot =sns.relplot(data=df[df.hierarchy=='[2]'], x="request_count", y="placements", style="k_means_seed", kind="line")
sns_plot.set(xlabel='Amount of VNF requests', ylabel='Sum of placed components')
sns_plot.savefig("figs/hierarchy_2_placements.pdf")

sns_plot =sns.relplot(data=df[df.hierarchy=='[2]'], x="request_count",ci='sd',markers=True, y="sum_total_delay", style="k_means_seed", kind="line", legend = False)
sns_plot.set(xlabel='Amount of VNF requests', ylabel='Sum of total delay of all requests in ms')

plt.legend(title='k-Means seed', loc='upper left', labels=['0', '1', '2', '3', '4'])
plt.ylim(0, 28)
plt.xticks(np.arange(1,6,1))
sns_plot.savefig("figs/hierarchy_2_sum_tot_delay.pdf")


sns_plot =sns.relplot(data=df[df.hierarchy=='[2]'], x="request_count", y="sum_e2e_delay", style="k_means_seed", kind="line")
sns_plot.savefig("figs/hierarchy_2_sum_e2e_delay.svg")
```
